## 3. Comparison of Remote and Non-Remote Job Salaries

### Objective
To compare the average salaries for remote and non-remote jobs. This analysis will determine if there are significant differences in compensation between these work settings.

### Methodology
- **Data Categorization**: Classify jobs into remote and non-remote categories based on the data.
- **Visualization**: Use bar charts to compare the average salaries of remote and non-remote jobs.


### Tools
- Python libraries: matplotlib, seaborn



```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Categorizing remote_ratio into three categories
data['remote_category'] = pd.cut(data['remote_ratio'], bins=[-1, 0, 99, 100], labels=['No Remote', 'Partial Remote', 'Fully Remote'])

# Setting the visual theme
sns.set(style="darkgrid")

# Comparing Remote and Non-Remote Job Salaries
plt.figure(figsize=(10, 5))  # Adjusted size to make it smaller
sns.barplot(
    x='remote_category', 
    y='salary_in_usd', 
    data=data, 
    estimator=np.mean, 
    errorbar=None  # Corrected from ci=None to errorbar=None
)
plt.title('Average Salary in USD by Remote Category')
plt.xlabel('Remote Work Category')
plt.ylabel('Average Salary (USD)')
plt.show()
```

The bar plot above illustrates the average salary in USD by remote work category. Here's what we can observe:

- **Fully Remote Jobs**: These positions tend to offer the highest average salaries among the categories analyzed. This suggests that roles allowing for full remote work command higher compensation.

- **Partial Remote and No Remote Jobs**: Both of these categories show lower average salaries when compared to fully remote positions. This indicates that less flexibility in remote work may be associated with lower salary offers.

### Implications

The observations suggest that jobs allowing full remote work might be associated with higher salary offers, potentially because these roles can target a wider geographical area. Furthermore, industries where remote work is feasible and valued might be more inclined to offer competitive salaries to attract and retain talent in a remote work environment.

The bar plot above displays the distribution of the top 10 job titles in the dataset. It illustrates the frequency of each job title, showing which positions are most common among the data entries. Jobs like "Data Scientist" and "Data Analyst" likely appear among the top due to their prevalent roles within the data science industry.

This visualization helps to identify the most in-demand roles and could be useful for understanding employment trends and prioritizing job positions within the data science sector.

## Data Preprocessing and Splitting

### Objective

Prepare the dataset for machine learning modeling by cleaning, encoding, and splitting the data into training and test sets.

### Steps of Preprocessing

1. **Dropping Unnecessary Columns**:
   - Columns like `employee_residence` and `company_location` were dropped as they are not essential for the salary prediction model.

2. **Encoding Categorical Variables**:
   - Categorical variables such as `experience_level`, `employment_type`, `job_title`, `salary_currency`, and `company_size` were encoded using `LabelEncoder` to convert them into numerical format suitable for modeling.

3. **Scaling the Data**:
   - The `remote_ratio` column was scaled using `StandardScaler` to normalize its distribution, ensuring that the scale of this feature matches other numerical features in the model.

### Data Splitting

- The dataset was split into features (`X`) and the target variable (`y`, which is `salary_in_usd`).
- Using `train_test_split` from `sklearn.model_selection`, the data was divided into training and test sets with an 80/20 split ratio, and a random state set for reproducibility.
